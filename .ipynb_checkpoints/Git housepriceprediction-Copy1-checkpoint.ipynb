```python
#RR
discrete_to_drop_cols = []

for features in discrete_numeric_features: 
    aa =(traindf[features].value_counts())
    aa1 = aa.apply(lambda x: (x/sum(aa.values)*100))
    print(aa1.head(3))    
    if aa1.head(1).values > 90:
        print('column to be dropped ')
        print('\n')
        discrete_to_drop_cols.append(features) 
```

```python

#RR
continuous_to_drop_cols = []

for features in continuous_numeric_features : 
    aa =(traindf[features].value_counts())
    aa1 = aa.apply(lambda x: (x/sum(aa.values)*100))
    print(aa1.head(5))
    if aa1.head(1).values > 85:
        print('column to be dropped ')
        print('\n')
        continuous_to_drop_cols.append(features) 
```

```python
#RR
print('continuous_to_drop_cols ',continuous_to_drop_cols)
print('discrete to dropcols    ' ,discrete_to_drop_cols)
```

```python
#RR
object_to_drop_cols = []

for features in Object_features : 
    aa =(traindf[features].value_counts())
    aa1 = aa.apply(lambda x: (x/sum(aa.values)*100))
    print(aa1.head(5))
    if aa1.head(1).values > 95:
        print('column to be dropped ')
        print('\n')
        object_to_drop_cols.append(features) 
```

```python
#RR
print('continuous_to_drop_cols ',continuous_to_drop_cols)
print('discrete to dropcols    ' ,discrete_to_drop_cols)
print('object_to_drop_cols     ' ,object_to_drop_cols)
```

```python
## Min max scalar
from sklearn.preprocessing import MinMaxScaler
mn = MinMaxScaler()
dfmn = mn.fit_transform(trainx)
```

```python
## convert the array to dataframe

trainx_scaled = pd.DataFrame(dfmn, columns=trainx.columns, index=trainx.index)
```

```python
#PCA
from sklearn.decomposition import PCA#Declared neeed library
pca_object = PCA(n_components=15)#19 features all total
principalComponents = pca_object.fit_transform(trainx)#fitting feature data for creating PCA
principalComponents
```

```python
principalDf = pd.DataFrame(data = principalComponents)#creating pandas dataframe from Principal componenets
principalDf
```

```python

import numpy as np
#The amount of variance that each PC explains
var= pca_object.explained_variance_ratio_
var
```

```python

#Cumulative Variance explains
var1=np.cumsum(np.round(pca_object.explained_variance_ratio_, decimals=4)*100)
var1
```
